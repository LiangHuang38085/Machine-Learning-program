一：分析背景

随着线上购物的普及程度日趋稳定，电商作为“开始进入者”的红利渐渐消失，用户的获取与保持难度也渐渐增大；在这种环境下，电商需要转换发展策略才能继续增长业务，在这个过程中通过与心理学的结合，利用好以往的购买数据显得尤为必要；通过数据分析可以制定更为有效的营销策略，也会帮助商家制定更好的生产计划及产品设计。

三：分析思路

数据分析的目的：
1）发现用户的行为规律

2）通过系统分析这些规律的不同维度得出关于商品本身畅销程度，淘宝网站的商品推荐精准度以及在各个环节用户转化与流失的一些结论

3）通过综合应用这些结论为商家、网站以及用户共同创造价值——即商家有了更准确的产品生产与设计规划，网站的推荐更准确从而带来了更大的流量和数据，用户能又快又准确的找到自己心仪的商品

针对以上数据分析的目的提出的问题：
不同时间跨度上的用户活跃度是怎样的？一周与一天内用户的行为规律是怎样的？

用户的复购率、复购商品有怎样的规律？用户复购畅销商品有哪些？

用户浏览的商品排名规律和购买的商品排名规律是否一致？（商品推荐是否精准？）

用户的转化与流失有哪些规律？（不同的购买流程是否导致了显著不同的购买转化率？）

以上分析将会得出怎样的结论以及如何综合应用这些结论来实现以上提出的数据分析目的？这些分析又有怎样的局限性？

七：结论总结及建议

总结

用户在一天的夜里零点到凌晨五点的时间段，活跃度逐渐降低；之后快速升高并于上午九点多趋于稳定直至下午六点多，在这期间购买行为在九点到十一点以及在十二点至十四点的增长幅度和速度值得注意；之后七点一直到晚上约十一点，pv、cart、buy、uv都有较大幅度增长；之后用户活跃度开始急剧下降，但绝对数量仍然较大，直到第二天的凌晨近一点才开始剧烈下降；用户活跃的高峰时间段是下午七点至晚上约十一点
工作日以及周末的活跃度需要更丰富的数据进一步分析
转化率最高的购买流程是点击-收藏-加购；其次是点击-加购；最后是点击-收藏
商品的推荐精准度不够，需要进一步提高
复购客户比例为54.4%，复购客户的购买次数占比为79.4%；复购客户比例较大，用户粘性较高，因此发展及保留客户的意义重大
畅销商品类目为'4756105'及‘3607361’；畅销的具体商品为‘812879’，‘3006793’，‘2331370’
建议

店家可以将活动促销，商品推荐，广告推送多放在每天的活跃高峰即七点到十一点，在每天的局部高峰适当布置资源；在低谷时段节省宣传资源
建议店家通过发放优惠券；建立会员制度（消费金额越大，会员卡等级越高，打折力度越大）；建立会员群，群内发放红包，也可以第一时间将新产品放到群里，通过和买家互动收集关于产品本身的反馈；通过这些方法以及其他的方式鼓励点击、收藏、加购来提高转化率
通过分析跨度更长的数据来甄别客户的等级，使用RFM分析方法针对不同的等级，使用有偿问卷、短信、邮件、店铺推送来提高客户的满意度及复购率
将畅销商品与非畅销商品进行捆绑优惠销售，可以在高峰时段进行有限数量的预售A/B测试，也可以在会员群内进行试验性销售，有助于店家淘汰不受欢迎的商品，以及精进产品设计及营销策略
